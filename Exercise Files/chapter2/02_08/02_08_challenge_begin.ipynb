### Code and data needed from previous videos

```python
medal = collections.namedtuple('medal',['year','athlete','team','event'])
medals = [medal(*line.strip().split('\t')) for line in open('goldmedals.txt','r')]
```

## Challenge solution

```python
medals_by_athletes = collections.Counter(medal.athlete for medal in medals)
# collections.Counter - counts the number of 'things' within a given tuple/list/data object
# In this case, it counts the times an athlete comes up for each medal in the medals list
```

```python
medals_by_athletes.most_common(5)
#.most_common() - returns the most common elements of a list
# The 5 returns the top 5 results
```

```python
# Collect events won by every athlete
events_by_athlete_set = collections.defaultdict(set)

for medal in medals:
    events_by_athlete_set[medal.athlete].add(medal.event)
```

```python
def howmany(tup):
    return len(tup[1])

sorted(events_by_athlete_set.items(),key=howmany,reverse=True)
```
